```python
import numpy as np
import matplotlib.pyplot as plt

def recursive_average(colors):
    """
    Compute the recursive average of the given list of colors.

    Parameters:
    - colors: List of color arrays

    Returns:
    - A new list of color arrays containing the recursive averages.
    """
    new_colors = [colors[0]]
    for i in range(1, len(colors)):
        avg_color = (colors[i] + new_colors[-1]) / 2.0
        new_colors.extend([avg_color, colors[i]])
    return new_colors

def generate_recursive_colors_v2(passes):
    """
    Generate a visualization of 2 random colors and their recursive averages for a number of passes.

    Parameters:
    - passes: Number of times to apply the recursive averaging process

    Returns:
    - An image showing the 2 endpoint colors and their recursive averages.
    """
    # Start with 2 random colors
    colors = [np.ones((10,10,3)) for _ in range(2)]
    for color in colors:
        color[:,:,0] = np.random.rand()
        color[:,:,1] = np.random.rand()
        color[:,:,2] = np.random.rand()

    # Apply the recursive averaging for the specified number of passes
    for _ in range(passes):
        colors = recursive_average(colors)

    # Horizontally stack the colors
    result = np.hstack(colors)

    return result
```

```python
import numpy as np                  #Import math functions from numpy
import matplotlib.pyplot as plt     #Import plot function from numpy

def recursive_average(colors):      #Computing recrusive average function
#with a list of colors

    new_colors = [colors[0]]    #A new list of colors containing
    for i in range(1, len(colors)):   #Recursive averages
        avg_color = (colors[i] + new_colors[-1]) / 2.0
        new_colors.extend([avg_color, colors[i]])
    return new_colors

def generate_recursive_colors_v2(passes, initial_colors):
  #Generating two random with their recursive averages for a number of passes
  #and their intial colors included

    colors = initial_colors.copy()

    for _ in range(passes):
        colors = recursive_average(colors)

    result = np.hstack(colors)      #Horizontally stacking the colors

    return result                 #Return function for the result

initial_colors = [np.ones((10, 10, 3)) for _ in range(2)]
for color in initial_colors:
    color[:,:,0] = np.random.rand()
    color[:,:,1] = np.random.rand()         #Number generators for the colors
    color[:,:,2] = np.random.rand()

passes_values = [0,1,2,3,4]   #Values for number of passes
imgs = [generate_recursive_colors_v2(p, initial_colors) for p in passes_values]

fig, axs = plt.subplots(1, len(passes_values), figsize=(15, 5))
#Creating a subplot

for ax, img, p in zip(axs, imgs, passes_values):
    ax.imshow(img)
    ax.axis('off')                      #Your plot settings
    ax.set_title(f'Passes={p}')

plt.tight_layout()
plt.show()            #Plotting the passes of colors below
```

```python
import numpy as np        #Import math functions from numpy
import matplotlib.pyplot as plt   #Import plot functions from numpy
import matplotlib.colors as mcolors   #Import colormaps and color formats from numpy
import matplotlib               #Import general functions from numpy
import imageio.v2 as imageio      #Import images from imageio
from PIL import Image              #Import image formating from PIL

def get_colors_lospec(url):       #Creating a color style "Lospec" with a url link
    im = imageio.imread(url)      #Reading images fro the url link
    rgb_list = (im/255)[0,:,0:3]   #Color scale format for the red green blue list
    float_list = list(np.linspace(0,1,len(rgb_list))) #Floating plot numbers
    #from the rgb_list
    cdict = dict()      #Python's dictionary
    for num, col in enumerate(['red', 'green', 'blue']):
      #A for loop that iterates a sequence of "red green and blue"
        col_list = [[float_list[i], rgb_list[i][num], rgb_list[i][num]] for i in range(len(float_list))]
        #Return function for the color list
        cdict[col] = col_list   #col_list is being put under Python's dictionary
    cmp = mcolors.LinearSegmentedColormap('my_cmp', segmentdata=cdict, N=256)
    #Creating colormap
    return cmp  #Return function for colormap
```
